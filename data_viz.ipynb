```python
# filter the AirQuality_station files
table_format_files_paths = [i for i in train_files_paths if 'AirQuality' not in i]

attributes = dict()
'''
    attibutes  = {
        "file_name1 " : [column1 , column2 , column3 ...]
        "file_name2 " : [column1 , column2 , column3 ...]
        ...
    }
'''
for file_path in table_format_files_paths:
    # get file name
    file_name = file_path.replace(train_dir , '')
    
    print ("Reading file : {}".format(file_name))
    df = pd.read_csv(file_path , encoding = "utf-8")
    
    # update each column to attribues
    attributes.update({file_name:list(df.columns.values)})
    
# show full column
pd.set_option('display.max_colwidth', -1)
pd.Series(attributes)
```

```python
# split table to devices base table
import csv
import os
import pandas 

path = '/root/data/pm25_data/datas/Taiwan_data/201703_Taiwan_sol.csv'
datas = dict()

with open(path, newline='', encoding = "utf-8") as csvfile:
    rows = csv.reader(csvfile)
    for i,row in enumerate(rows):
        if i == 0 : continue
        data = [a.strip() for a in row]
        dev = data[2]
        if dev in datas:
            datas[dev].append(data)
        else:
            datas.update({dev : [data]})
print ('Finish , total dev : {}'.format(len(datas.keys())))

for keys , values in datas.items():
    out_path = '/root/data/pm25_data/datas/sensors/test/'
#     if len(values ) > 1000:
    with open(out_path + keys + ".csv", "a", encoding = "utf-8") as f:
        writer = csv.writer(f)
        writer.writerows(values)

            
```

```python
# linear interpolate each time table

file_names = os.listdir(out_path)

for file_name in file_names:
    test_file = os.path.join(out_path , file_name)
    print (test_file)
    def parser(x):
        
        return pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
    
    series = pd.read_csv(test_file, header=None, parse_dates=[[0,1]], index_col=0, squeeze=True, date_parser=parser)

    upsampled = series.resample('5T').mean().interpolate('linear')

    upsampled.to_csv(test_file, encoding='utf-8',header = None )

print ('finish {} files'.format(len(file_names)))
```
